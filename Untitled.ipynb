```python
import dpkt
import socket
import datetime
def printPcap(pcap):
    try:
        for timestamp, buf in pcap:
            eth = dpkt.ethernet.Ethernet(buf) #获得以太包，即数据链路层包
            print("ip layer:"+eth.data.__class__.__name__) #以太包的数据既是网络层包
            print("tcp layer:"+eth.data.data.__class__.__name__) #网络层包的数据既是传输层包
            #print("http layer:" + eth.data.data.data.__class__.__name__) #传输层包的数据既是应用层包
            print('Timestamp: ',str(datetime.datetime.utcfromtimestamp(timestamp+ 28800))) #打印出包的抓取时间
            if not isinstance(eth.data, dpkt.ip.IP):
                print('Non IP Packet type not supported %s' % eth.data.__class__.__name__)
                continue
            ip = eth.data
            do_not_fragment =bool(ip.off & dpkt.ip.IP_DF)
            more_fragments =bool(ip.off & dpkt.ip.IP_MF)
            fragment_offset = ip.off & dpkt.ip.IP_OFFMASK
            print('P: %s -> %s (len=%d ttl=%d DF=%d MF=%d offset=%d)' % (socket.inet_ntoa(ip.src), 
                                                                     socket.inet_ntoa(ip.dst), 
                                                                     ip.len, 
                                                                     ip.ttl, 
                                                                     do_not_fragment, 
                                                                     more_fragments,
                                                                     fragment_offset))
    except:
        pass
def main():
    f =open('1hour.pcap','rb')
    pcap = dpkt.pcap.Reader(f)
    printPcap(pcap)
    print(pcap)
if __name__ =='__main__':
    main()
    
```

```python
import dpkt
import socket
import datetime
def printPcap(pcap):
    try:
        for timestamp, buf in pcap:
            eth = dpkt.ethernet.Ethernet(buf) #获得以太包，即数据链路层包
            #print("ip layer:"+eth.data.__class__.__name__) #以太包的数据既是网络层包
            #print("tcp layer:"+eth.data.data.__class__.__name__) #网络层包的数据既是传输层包
            #print("http layer:" + eth.data.data.data.__class__.__name__) #传输层包的数据既是应用层包
            
            if eth.data.__class__.__name__=='ARP' or eth.data.data.__class__.__name__=='IGMP' or eth.data.data.__class__.__name__=='IP6':
                continue
            if not isinstance(eth.data, dpkt.ip.IP):
                #print('Non IP Packet type not supported %s' % eth.data.__class__.__name__)
                continue
                
           # print('Timestamp: ',str(datetime.datetime.utcfromtimestamp(timestamp+ 33600))) #打印出包的抓取时间
            
            ip = eth.data
            do_not_fragment =bool(ip.off & dpkt.ip.IP_DF)
            more_fragments =bool(ip.off & dpkt.ip.IP_MF)
            fragment_offset = ip.off & dpkt.ip.IP_OFFMASK
            file = 'port.txt'
            if eth.data.data.__class__.__name__=='TCP':
                #pass
                #print(eth.data.data.__class__.__name__,eth.data.data.sport)
                s = '1'+','+str(eth.data.data.sport)+'\n'
                with open(file, 'a') as f:
                    f.write(s)
                #print("dport:",eth.data.data.dport)
            elif eth.data.data.__class__.__name__=='UDP':
                s = '2'+','+str(eth.data.data.sport)+'\n'
                with open(file, 'a') as f:
                    f.write(s)
            else:
                #print(eth.data.data.__class__.__name__,0)
                s = '3'+','+'0'+'\n'
                with open(file, 'a') as f:
                    f.write(s)
    except:
        pass
def main():
    f =open('1hour.pcap','rb')
    pcap = dpkt.pcap.Reader(f)
    printPcap(pcap)
    #print(pcap)
if __name__ =='__main__':
    main()
    
```
